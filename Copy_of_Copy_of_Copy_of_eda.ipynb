```python
data['type'] = data['type'].replace(['Industrial Area','Industrial Areas'],'Industrial')
data['type'] = data['type'].replace(['Sensitive Area','Sensitive Areas'],'Sensitive')
data['type'] = data['type'].replace(['Residential, Rural and other Areas','Residential and others'],'Residential')
```

# Data Visualization

```python
#barplot for each type
data['type'].value_counts().plot(kind = 'bar')
```

```python
data[['so2', 'state']].groupby(['state']).median().sort_values("so2", ascending = False).plot.bar(figsize=(20,10))
plt.show()
```

from the above figure, we see that so2 level is highest in Uttaranchal and lowest in Himachal pradesh.


```python
plt.subplots(figsize=(20,10))
sns.barplot(x='state',y = 'no2',palette = 'crest',data= data)
plt.xticks(rotation = 70)
plt.title('level of no2 in states',fontsize=30)
plt.xlabel('state',fontsize=20)
plt.ylabel('no2',fontsize=20)
plt.show()
```

It is clear from the figure that West Bengal has the maximum level of no2, whereas Arunachal Pradesh has the least.

```python
plt.subplots(figsize=(20,10))
sns.barplot(x='state',y = 'rspm',palette = 'RdPu',data = data)
plt.xticks(rotation = 70)
plt.title('level of rspm in states',fontsize=30)
plt.xlabel('state',fontsize=20)
plt.ylabel('rspm',fontsize=20)
plt.show()
```

From the above plot, we see that Delhi has the highest concentration of rspm

```python
plt.subplots(figsize=(20,10))
sns.barplot(x='state',y = 'spm',palette = 'rocket',data = data)
plt.xticks(rotation = 70)
plt.title('level of spm in states',fontsize=30)
plt.xlabel('state',fontsize=20)
plt.ylabel('spm',fontsize=20)
plt.show()
```

We see that Uttar Pradesh(UP) and Delhi pair again tops the list. UP and Delhi have a comparable concentration of spm as well as rspm.
